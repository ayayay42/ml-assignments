```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

x_train = np.load('/content/drive/MyDrive/a.Unif/MASTER1/ML/A4_train.npz')['images']
y_train = np.load('/content/drive/MyDrive/a.Unif/MASTER1/ML/A4_train.npz')['labels']
x_test = np.load('/content/drive/MyDrive/a.Unif/MASTER1/ML/A4_test.npz')['images']
y_test = np.load('/content/drive/MyDrive/a.Unif/MASTER1/ML/A4_test.npz')['labels']

# Convert string labels as int, followed by one-hot encoding
labels = sorted(list(set(y_train)))
y_train = keras.utils.to_categorical([labels.index(x) for x in y_train])
y_test = keras.utils.to_categorical([labels.index(x) for x in y_test])
```

```python
from tensorflow.keras import layers
from tensorflow.keras.initializers import RandomNormal

# Define the neural network model correctly
model = keras.Sequential([
    keras.Input(shape=(32, 128)),  # Explicit Input layer
    layers.Flatten(),  # Flatten layer
    layers.Dense(15, activation='softmax', kernel_initializer=RandomNormal(), bias_initializer=RandomNormal())
])

# Compile the model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['categorical_accuracy']
)

model.fit(x_train, y_train, epochs=100, batch_size=32)

test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

print(model.summary())

model.save('/content/drive/MyDrive/a.Unif/MASTER1/ML/model_q_2.keras')
```

```python
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

print(model.summary())
```

```python
# List of model configurations
configurations = [
    (64, 64), (64, 256), (256, 64), (256, 256), (1024, 256),
    (256, 1024), (64, 64, 64), (256, 256, 256)
]

results = []

for idx, config in enumerate(configurations):
    model = keras.Sequential()
    model.add(keras.Input(shape=(32, 128)))
    model.add(layers.Flatten())

    # Add hidden layers
    for units in config:
        model.add(layers.Dense(units, activation='relu', kernel_initializer=RandomNormal(), bias_initializer=RandomNormal()))

    # Output layer
    model.add(layers.Dense(15, activation='softmax', kernel_initializer=RandomNormal(), bias_initializer=RandomNormal()))

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=1e-5),
        loss='categorical_crossentropy',
        metrics=['categorical_accuracy']
    )

    # Train and evaluate model
    accuracies = []
    for _ in range(3):  # First do 3 runs for initial evaluation
        model.fit(x_train, y_train, epochs=100, batch_size=32, verbose=0)
        _, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
        accuracies.append(test_accuracy)

    mean_test_acc = np.mean(accuracies)
    num_params = model.count_params()
    results.append((idx, mean_test_acc, num_params))
    print(f"Config {idx}: {config}, Mean Test Accuracy: {mean_test_acc:.3f}, Parameters: {num_params}")

# Find the best configuration
best_config = max(results, key=lambda x: x[1])
print(f"Best Configuration: {best_config[0]}, Accuracy: {best_config[1]:.3f}, Parameters: {best_config[2]}")
```
